```python
distortion = []
c_labels = []
```

```python
vehicle_df_z = vehicle_df.drop('car name', axis=1)
```

```python
from sklearn.cluster import KMeans
cluster_range = range(1, 10)
for num_clusters in cluster_range:
  clusters = KMeans( num_clusters, n_init = 3)
  clusters.fit(vehicle_df_z)
  labels = clusters.labels_
  centroids = clusters.cluster_centers_
  distortion.append( clusters.inertia_ ) 
```

```python
clusters_df = pd.DataFrame( { "num_clusters":cluster_range, "cluster_errors": distortion} )
clusters_df[0:15]
```

```python
plt.figure(figsize=(12,6))
plt.plot( clusters_df.num_clusters, clusters_df.cluster_errors, marker = "o" )
```

```python
cluster = KMeans( n_clusters = 3, random_state = 2354 )
cluster.fit(vehicle_df_z)
vehicle_df_z_copy = vehicle_df_z.copy(deep = True)
```

```python
centroids = cluster.cluster_centers_
centroids
```

```python
centroid_df = pd.DataFrame(centroids, columns = list(vehicle_df_z) )
centroid_df
```

```python
prediction=cluster.predict(vehicle_df_z)
vehicle_df_z["GROUP"] = prediction
```

```python
vehicle_df_z['labels'] = cluster.labels_
```

```python
vehicle_df_z.groupby(["labels"]).count()
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(vehicle_df_z['GROUP'],vehicle_df_z['mpg'], c=vehicle_df_z['labels'], cmap='prism')
plt.show()
```
