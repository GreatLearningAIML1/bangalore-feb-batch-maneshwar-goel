```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
sc = StandardScaler()
X_std =  sc.fit_transform(X)          
cov_matrix = np.cov(X_std.T)
print('Covariance Matrix \n%s', cov_matrix)
```

```python
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
print('Eigen Vectors \n%s', eigenvectors)
print('\n Eigen Values \n%s', eigenvalues)
```

```python
eig_pairs = [(eigenvalues[index], eigenvectors[:,index]) for index in range(len(eigenvalues))]
eig_pairs.sort()
eig_pairs.reverse()
print(eig_pairs)
eigvalues_sorted = [eig_pairs[index][0] for index in range(len(eigenvalues))]
eigvectors_sorted = [eig_pairs[index][1] for index in range(len(eigenvalues))]
print('Eigenvalues in descending order: \n%s' %eigvalues_sorted)
```

```python
tot = sum(eigenvalues)
var_explained = [(i / tot) for i in sorted(eigenvalues, reverse=True)]
cum_var_exp = np.cumsum(var_explained)
```

```python
plt.bar(range(1,7), var_explained, alpha=0.5, align='center', label='individual explained variance')
plt.step(range(1,7),cum_var_exp, where= 'mid', label='cumulative explained variance')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal components')
plt.legend(loc = 'best')
plt.show()
```

```python
X_centered = X - X.mean(axis=0)
```

```python
pca = PCA(n_components=3)
pca.fit(X_centered)
```

```python
X_pca = pca.transform(X_centered)
```

```python
# Applying K-Means Clustering to cluster the PCA data.
```

```python
from sklearn.cluster import KMeans
labels = KMeans(n_clusters=3).fit_predict(X_pca)
```

```python
for label, color in zip(range(3), ['r', 'g', 'b']):
    plt.scatter(X_pca[labels == label, 0], X_pca[labels == label, 1], c=color)
plt.show()
```

```python
new_df = pd.DataFrame({'Pc1':X_pca[:,0],'Pc2':X_pca[:,1],'Pc3':X_pca[:,2]})
```
